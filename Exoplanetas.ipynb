```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import figure
```

```python
csv_exoplanetas = pd.read_csv('oec.csv')
```

```python
csv_exoplanetas.shape
```

```python
exoplanets_discovered = csv_exoplanetas[['PlanetIdentifier' , 'DiscoveryYear','DiscoveryMethod']]
```

```python
##Eliminar planetas no descubiertos (eliminar filas)
exoplanets_discovered = exoplanets_discovered[exoplanets_discovered['DiscoveryMethod'].notna()]
```

```python
# Gráfica planetas descubiertos por año
planeta_por_anio= exoplanets_discovered['DiscoveryYear'].groupby([exoplanets_discovered['DiscoveryYear']]).count()
y = planeta_por_anio
x = planeta_por_anio.index
figure(figsize=(12, 8))
plt.grid()
plt.xlabel("Año")
plt.ylabel("Cantidad")
plt.title("Exoplanetas descubiertos por año")
xlabels = planeta_por_anio.index
plt.plot(x, y, color="blue")
```

```python
exoplanets_discovered['DiscoveryMethod'].groupby([exoplanets_discovered['DiscoveryYear'], exoplanets_discovered['DiscoveryMethod']]).count()
```

```python
metodo_descubrimiento = exoplanets_discovered.groupby(['DiscoveryYear', 'DiscoveryMethod'])[['DiscoveryMethod']].agg(['count'])
```

```python
metodo_descubrimiento.unstack().plot(kind='bar',
                                     stacked='False',
                                     figsize=(16, 10),
                                    fontsize=12,
                                    label='test')
plt.xlabel('Año',fontsize=14)
plt.ylabel('Cantidad',fontsize=14)
plt.title('Exoplanetas descubiertos por año (MÉTODOS)',fontsize=14)
plt.legend(loc='upper left', prop={'size': 14})
```
